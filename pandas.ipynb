```python
#verifier s'il y a des donnees manquantes
df.isna().sum()
```

```python
#suppression les lignes ou il y a des donnees maquantes
df.dropna()
```

```python
#Recodage de sexe (0 pour masculin et 1 pour féminin)
df['sexe']=df['sexe'].map({'masculin':0,'feminin':1})
```

```python
#supprimer les variable depression, vaisseau et pic du dataframe
d = df.drop(['depression','vaisseau','pic'],axis=1)
```

```python
#Representation graphique de la variable coeur
plt.hist(df[['coeur']], color="gray",edgecolor='white')
plt.title("Histogamme de la variable 'AGE'", fontsize=20)
plt.xlabel("coeur")
```

```python
#Representation graphique de toutes les variables qualitatives
list_qual = df.select_dtypes(include = 'object')

plt.figure(figsize=(30, 15))
plt.gcf().subplots_adjust(wspace=1, hspace=1)
i=1

for var in list_qual:
    plt.subplot(2,3,i)
    x=df[var].value_counts()
    x.plot(kind='bar')
    plt.title(var, fontsize=40)
    plt.xticks(fontsize=20, rotation=45)
    plt.yticks(fontsize=20)

    i+=1
```

```python
#Representation graphique de toutes les variables quantitative
list_quant = df.select_dtypes(np.number)

for i in list_quant:
    plt.hist(df[i], edgecolor='black', label=i)
    plt.title(i)
    plt.legend()
    plt.show()
```
